# Doppler acceleration for a transmitter orbiting Proxima Centauri
---
The goal of this notebook is to predict the doppler acceleration of a transmitter on Prox Cen B, orbiting Prox Cen.

```python
import numpy as np
import astropy as ap
import pylab as plt
from astropy import units as u
```

```python
%matplotlib inline
```

The maximum Doppler drift due to a body's rotation is given by:
$$
\dot{\nu} = \frac{4 \pi^2 R}{P^2} \frac{ \nu_0 }{c}
$$ 
where $P$ is rotational/orbital period, $R$ is body/orbit radius, and $c$ is the speed of light. <br/>
Data below is taken from [Revisting Proxima with ESPRESSO](https://arxiv.org/pdf/2005.12114)

```python
P = 11.18427 * u.day
R = 0.0485 * u.AU
c = 2.998*10**8 * u.m/u.s
print("P = %s \nR = %s \nc = %s"%(P,R,c))
```

```python
dopp = lambda nu0: 4*np.pi**2*R/(c*P**2)*nu0
#The search range will be from 700 MHz to 4 GHz
nu = np.arange(700*10**3,4*10**6,100)
plt.plot(nu,dopp(nu), label='Expected Doppler Shift')
plt.grid()
plt.title('Maximum Expected Doppler Drift for a Reciever on \nProxima Centauri b')
plt.xlabel('Emitted Frequency (MHz)')
plt.ylabel('Doppler Shift (Hz/s)')
plt.show()
```
